### SVM as a regrasion model

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,median_absolute_error
data3=pd.read_csv("Position_Salaries.csv")
data3.drop('Position',axis=1,inplace=True)
x=data3.drop('Salary',axis=1)
y=data3['Salary']

#scaling the data
xscaler=StandardScaler()
yscaler=StandardScaler()
x=xscaler.fit_transform(x)


plt.scatter(x,y)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(x,y)
y_pred= regressor.predict(x)
print('mean sq eroor',mean_squared_error(y,y_pred))
print('mean abs error',mean_absolute_error(y,y_pred))
print('mean abs error',median_absolute_error(y,y_pred))

plt.scatter(x,y_pred.reshape(-1,1),color='red')
plt.plot(x,y_pred.reshape(-1,1),color='blue')

```
