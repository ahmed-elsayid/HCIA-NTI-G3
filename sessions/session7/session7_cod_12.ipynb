
# Decision Tree



```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
from sklearn import metrics



```

 ### reading data

```python
columns_names=['Pregnancies','Glucose','BP','SkinThickness','Insulin','BMI','pedigree','Age','label']
data=pd.read_csv('diabetes.csv',names=columns_names)
data.head()
```

### x, y and split the data 

```python
y=data['label'].values.reshape(-1,1)
x=data.drop('label',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### creat the model

```python
dessiontree_model=DecisionTreeClassifier(criterion="entropy",max_depth=3)
dessiontree_model.fit(x_train,y_train)
y_pred=dessiontree_model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

### testes

```python
print("Confusion Matrix:",metrics.confusion_matrix(y_test,y_pred))
print("Classification Report:",metrics.classification_report(y_test,y_pred))
```

### plot

```python
plt.figure(figsize=(20,20))
plot_tree(dessiontree_model,filled=True,rounded=True,feature_names=x.columns,class_names=['0','1'])
plt.savefig("DecisionTree.png")
```
